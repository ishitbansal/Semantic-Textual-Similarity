```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
### Mono-Lingual Analysis
data = {
    'Model': ['Word2Vec', 'Word2Vec using Linear Regression', 'Word2Vec using GRU Regression', 
              'Word2Vec using BiLSTM Regression', 'Word2Vec using BiLSTM Attention', 
              'BERT (untrained)', 'BERT (Fine-Tuned)', 'RoBERTa (untrained)', 
              'RoBERTa (Fine-Tuned)', 'Doc2Vec (untrained)', 'Doc2Vec (BiLSTM)', 
              'Universal Sentence Encoder (untrained)', 'Universal Sentence Encoder (Fine-tuned)', 
              'Sentence Transformers (untrained)', 'Sentence Transformers (Fine-Tuned)', 
              'Siamese BiLSTM Model'],
    'Train': [0.68, 0.47, 0.75, 0.77, 0.77, 0.13, 0.90, 0.17, 0.95, 0.50, 0.87, 0.70, 0.88, 0.78, 0.78, 0.22],
    'Validation': [0.73, 0.16, 0.40, 0.36, 0.37, -0.14, 0.83, -0.09, 0.91, 0.61, 0.38, 0.74, 0.55, 0.82, 0.82, 0.12],
    'Test': [0.63, 0.19, 0.37, 0.34, 0.37, 0.16, 0.82, -0.02, 0.91, 0.54, 0.34, 0.66, 0.54, 0.76, 0.76, 0.12]
}

df = pd.DataFrame(data)
df_melted = pd.melt(df, id_vars='Model', var_name='Dataset', value_name='Pearson Coefficient')

plt.figure(figsize=(12, 8))
sns.barplot(x='Model', y='Pearson Coefficient', hue='Dataset', data=df_melted)
plt.xticks(rotation=90)
plt.title('Pearson Coefficient for Different Models Across Datasets')
plt.xlabel('Model')
plt.ylabel('Pearson Coefficient')
plt.legend(title='Dataset')
plt.tight_layout()
plt.show()
```

```python
df_transposed = df.set_index('Model').T.reset_index().rename(columns={'index': 'Dataset'})
df_melted = pd.melt(df_transposed, id_vars='Dataset', var_name='Model', value_name='Pearson Coefficient')

plt.figure(figsize=(12, 8))
sns.barplot(x='Dataset', y='Pearson Coefficient', hue='Model', data=df_melted)
plt.xticks(rotation=90)
plt.title('Pearson Coefficient for Different Datasets Across Models')
plt.xlabel('Dataset')
plt.ylabel('Pearson Coefficient')
plt.legend(title='Model', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Model', y='Train', data=df)
plt.xticks(rotation=90)
plt.title('Pearson Coefficient for Train Dataset')
plt.xlabel('Model')
plt.ylabel('Pearson Coefficient')
plt.ylim(0, 1)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x='Model', y='Validation', data=df)
plt.xticks(rotation=90)
plt.title('Pearson Coefficient for Validation Dataset')
plt.xlabel('Model')
plt.ylabel('Pearson Coefficient')
plt.ylim(0, 1)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x='Model', y='Test', data=df)
plt.xticks(rotation=90)
plt.title('Pearson Coefficient for Test Dataset')
plt.xlabel('Model')
plt.ylabel('Pearson Coefficient')
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```
